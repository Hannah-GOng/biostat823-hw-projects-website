### Interactive Visulization1: Years vs Number of Incident

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="year", value = 2000, clearable = False,
        options=[
            {'label': x, 'value': x}
            for x in df['Year'].unique()
        ]),
    dcc.Graph(id = 'graph', figure = {})
])
    

@app.callback(
    Output("graph", "figure"), 
    [Input("year", "value")])
def display(year):
    df_year = df[df['Year'] == year]
    
    data = dict(type='choropleth', 
            locations = df_year['Code'], 
            z = df_year['Incidence of malaria (per 1,000 population at risk) (per 1,000 population at risk)'], 
            text = df_year['Entity'])
    
    layout = dict(title = 'Incidence of malaria (per 1,000 population at risk) (per 1,000 population at risk',
              geo = dict(projection = {'type':'robinson'},
                         showlakes = True, 
                         lakecolor = 'rgb(0,191,255)'))
    
    fig = go.Figure(data = [data], 
              layout = layout)

    return fig
    

app.run_server(debug=True, use_reloader=False) 
```

### Interactive Visulization2: Age Distribution vs Number of Death in each country Over years

```python
df2 = pd.read_csv('malaria_deaths_age.csv')
df_under5 = df2[df2['age_group'] == 'Under 5']
```

```python
df_under5.head()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="country", value = 'Afghanistan', clearable = False,
        options=[
            {'label': x, 'value': x}
            for x in df_under5['entity'].unique()
        ]),
    dcc.Graph(id = 'graph', figure = {})
])
    

@app.callback(
    Output("graph", "figure"), 
    [Input("country", "value")])

def display(country):
    df_country = df_under5[df_under5['entity'] == country]
    
    fig = go.Figure(data=go.Bar(x = df_country['year'],
                                y = df_country['deaths']))
    
    fig.update_layout(title_text='Death of age group under 5')

    return fig
    

app.run_server(debug=True, use_reloader=False) 
```

### Interactive Visulization3: Years vs Number of Death per 100,000 people

```python
df3 = pd.read_csv('malaria_deaths.csv')
```

```python
df3.head()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Slider(
        id='year',
        min=df3['Year'].min(),
        max=df3['Year'].max(),
        value=df3['Year'].min(),
        marks={str(year): str(year) for year in df3['Year'].unique()},
        step=None
    ),
    dcc.Graph(id = 'graph', figure = {})
])
    

@app.callback(
    Output("graph", "figure"), 
    [Input("year", "value")])

def display(year):
    df_year = df3[df3['Year'] == year]
    
    data = dict(type='choropleth', 
            locations = df_year['Code'], 
            z = df_year['Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)'], 
            text = df_year['Entity'])
    
    layout = dict(title = 'Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)',
              geo = dict(projection = {'type':'robinson'},
                         showlakes = True, 
                         lakecolor = 'rgb(0,191,255)'))
    
    fig = go.Figure(data = [data], 
              layout = layout)

    return fig
    

app.run_server(debug=True, use_reloader=False) 
```
